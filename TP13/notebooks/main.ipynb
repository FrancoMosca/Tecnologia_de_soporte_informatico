```python
import pandas as pd
import matplotlib.pyplot as gr
import matplotlib.pyplot as gr2
import matplotlib.pyplot as gr3
import matplotlib.pyplot as gr4

df = pd.read_csv("../data/analisisTaller.csv")

gr.title('Hasta qu mes entregaste trabajos?')
x = df['Hasta qu mes entregaste trabajos?'].unique()
y = df['Hasta qu mes entregaste trabajos?'].value_counts().tolist()
gr.bar(x,y)
gr.xlabel('Meses')
gr.ylabel('Cant de alumnos')
gr.xticks(rotation=45)
gr.show()


gr2.title('Tu cumplimiento con las entregas tiene que ver con que ests en el ltimo ao y quers terminar?')
cant = [df["Tu cumplimiento con las entregas tiene que ver con que ests en el ltimo ao y quers terminar?"].value_counts()["S"],df["Tu cumplimiento con las entregas tiene que ver con que ests en el ltimo ao y quers terminar?"].value_counts()["No"]]
sino = ["S","No"]
gr2.pie(cant, labels=sino, autopct='%1.1f%%')
gr2.show()

gr3.title('Cunto te gusta Python como lenguaje de programacin?')
x = df['Cunto te gusta Python como lenguaje de programacin?'].unique()
y = df['Cunto te gusta Python como lenguaje de programacin?'].value_counts().tolist()
gr3.bar(x,y)
gr3.xlabel('Gusto')
gr3.ylabel('Cant de alumnos')
gr3.xticks(rotation=45)
gr3.show()

gr4.title('Pudo entregar todos los trabajos de la materia?')
cant = [df["Pudo entregar todos los trabajos de la materia?"].value_counts()["S"],df["Pudo entregar todos los trabajos de la materia?"].value_counts()["No"]]
sino = ["S","No"]
gr4.pie(cant, labels=sino, autopct='%1.1f%%')
gr4.show()
```

### Anlisis de la materia

Viendo los graficos, nos damos cuenta que agosto fue el mes del ao en el que mas alumnos dejaron de estar al dia con sus trabajos,
y que al ser el ultimo ao de escuela muchos tambien se relajaron y no prestaron la atencion debida. Esto sumado a que no hay un particular interes
por aprender el lenguaje utilizado en estos trabajos. Se ve reflejado en la cantidad de gente que no pudo terminar sus trabajos en tiempo y forma.
