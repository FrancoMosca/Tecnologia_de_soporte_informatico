```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2
import numpy as np

data = pd.read_csv("../data/personal_estrategico_vacunado_covid.csv")
data = data.loc[0:100]
data=data.dropna()
```

```python

x = data['Cargo'].unique()
y = data['Cargo'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Cargos')
gf1.xlabel('Cargos')
gf1.ylabel('Cantidad')
gf1.xticks(rotation=90,fontsize=11)
gf1.show()
```

```python
xx = data['Autorizado_Por'].unique()
yy = data['Autorizado_Por'].value_counts().tolist()

gf2.plot(xx, yy)
gf2.title('Autorizado_Por')
gf2.xlabel('Autorizado_Por')
gf2.ylabel('Cantidad')
gf2.xticks(rotation=90,fontsize=11)
gf2.show()
```
