```python
import pandas as pd
import matplotlib.pyplot as gf1
import matplotlib.pyplot as gf2


data = pd.read_csv("../data/Instrumentos.csv")
```

```python
x = data['¿Cual?'].value_counts().index.tolist()
y = data['¿Cual?'].value_counts().tolist()

gf1.bar(x,y)
gf1.title('Instrumento mas tocado')
gf1.xlabel('Instrumento')
gf1.ylabel('Cantidad de personas que toca ese instrumento')
gf1.xticks(rotation=90,fontsize=12)
gf1.show()
```

```python
xx = data['¿Tocas_algun_instrumento/cantas?'].unique()
yy = data['¿Tocas_algun_instrumento/cantas?'].value_counts().tolist()

gf2.bar(xx,yy)
gf2.title('Gente que toca instrumentos')
gf2.xlabel('¿toca o no?')
gf2.ylabel('Cantidad de personas que tocan')
gf2.xticks(rotation=90,fontsize=12)
gf2.show()
```

Llegue a la conclucion que hay mas gente que toca instrumentos que la que no, y que el instrumento mas tocado es la guitarra
